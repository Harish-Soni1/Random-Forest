```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
%matplotlib inline
```

```python
salary=pd.read_csv("E:\\itsstudytym\\assignments\\Naive Bayes\\SalaryData_Train.csv")
salary.head()
```

```python
salary=salary.drop(['educationno','relationship','native','maritalstatus','race','sex'],axis=1)
salary.head()
```

```python
salary.isnull().sum()
```

```python
for i in range(0,len(salary['Salary'])):
    if salary['Salary'][i]==' <=50K':
        salary['Salary'][i]=0
    else:
        salary['Salary'][i]=1
        
```

```python
salary.head(20)
```

```python
workclass={
    ' State-gov':1,
    ' Federal-gov':2,
    ' Local-gov':3,
    ' Private':4,
    ' Self-emp-inc':5,
    ' Self-emp-not-inc':6,
    ' Without-pay':7
}

education={
    ' Doctorate':1,
    ' Prof-school':2,
    ' Masters':3,
    ' Bachelors':4,
    ' Some-college':5,
    ' HS-grad':6,
    ' Assoc-acdm':7,
    ' Assoc-voc':8,
    ' 12th':9,
    ' 11th':10,
    ' 10th':11,
    ' 9th':12,
    ' 7th-8th':13,
    ' 5th-6th':14,
    ' 1st-4th':15,
    ' Preschool':16
}
```

```python
salary['education']=salary['education'].map(education)
salary['workclass']=salary['workclass'].map(workclass)
salary.head()
```

```python
Occupation=pd.get_dummies(salary['occupation'],drop_first=True)
Occupation
```

```python
salary['Salary']=salary['Salary'].astype(int)
```

```python
X=salary.iloc[:,:-1]
y=salary['Salary']
```

```python
X=pd.concat([X,Occupation],axis=1)
X.head()
```

```python
X=X.drop('occupation',axis=1)
X.head()
```

```python
X.dtypes
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC=RandomForestClassifier()
```

```python
RFC.fit(X_train,y_train)
RFC.score(X_test,y_test)
```
